## Переведем аннотацию и получим файлы промоторов/экзонов/межгенников/интронов и downstream

```python
def calculate_scaffold_lengths(fasta_file, output_file):
    with open(output_file, 'w') as out:
        for record in SeqIO.parse(fasta_file, "fasta"):
            scaffold_name = record.id
            scaffold_length = len(record.seq)
            out.write(f"{scaffold_name}\t{scaffold_length}\n")
calculate_scaffold_lengths('GCA_036169595.1_ASM3616959v1_genomic.fna', 'scaffolds.sizes')
```

```python
!awk '{if($1 !~ /^#/ ) print $1"\t"$4-1"\t"$5"\t"$3"\t"$10"\t"$7}' GCA_036169595.1_ASM3616959v1_genomic.gtf > annotation.bed
```

```python
from Bio import SeqIO

input_fasta = 'GCA_036169595.1_ASM3616959v1_genomic.fna'
output_dir = 'scaffolds/'

def split_fasta_by_scaffolds(input_fasta, output_dir):
    fasta_sequences = SeqIO.parse(open(input_fasta),'fasta')
    for fasta in fasta_sequences:
        name, sequence = fasta.id, str(fasta.seq)
        with open(output_dir + name + ".fasta", "w") as output_file:
            output_file.write(f">{name}\n{sequence}\n")

split_fasta_by_scaffolds(input_fasta, output_dir)
```

```python
import pandas as pd
import os
directory = "scaffolds"
results = []
for filename in os.listdir(directory):
    if filename.endswith(".Z-SCORE"):
        clean_filename = filename.rstrip('.Z-SCORE')
        filepath = os.path.join(directory, filename)
        df = pd.read_table(filepath, delim_whitespace=True, skiprows=1,
                           names=['Start', 'End', 'None1', 'None2', 'None3', 'Z-score', 'Sequence', 'None4'])
        filtered_df = df[df['Z-score'] >= 400]
        for index, row in filtered_df.iterrows():
            results.append((clean_filename, row['Start'], row['End']))
            print((clean_filename, row['Start'], row['End']))
print(results)
```

```python
zdna = pd.DataFrame(results, columns=['Chromosome', 'Start', 'End'])
zdna['Chromosome'] = zdna['Chromosome'].str.replace('.fasta', '', regex=False)
zdna
```

```python
from fuc import pybed
bf = pybed.BedFrame.from_frame(meta=[], data=zdna)
bf.to_file('zhunt.bed')
```
